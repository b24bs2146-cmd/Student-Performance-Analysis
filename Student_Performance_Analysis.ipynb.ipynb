```
import pandas as pd

df = pd.read_csv("student_performance.csv")
df.head()
```

```
df.describe()
```

```
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(7,5))
sns.scatterplot(x="study_hours", y="math_score", data=df)
plt.title("Study Hours vs Math Score")
plt.xlabel("Study Hours")
plt.ylabel("Math Score")
plt.show()
```

### Insight 1: Study Hours vs Math Score

Students who study for more hours per day generally tend to score higher in mathematics.
this indicates a positive relationship between study time and academic performance.

```
df["average_score"] = (df["math_score"] + df["reading_score"] + df["writing_score"]) / 3
df.head()
```

```
plt.figure(figsize=(7,5))
sns.scatterplot(x="sleep_hours", y="average_score", data=df)
plt.title("Sleep Hours vs Average Score")
plt.xlabel("Sleep Hours")
plt.ylabel("Average Score")
plt.show()
```

### Insight 2: Sleep Hours vs Academic Performance

Students who get around 78 hours of sleep tend to have higher average scores.
This suggests that adequate sleep plays an important role in academic performance.


```
df.groupby("gender")["average_score"].mean()
```

```
plt.figure(figsize=(7,5))
sns.barplot(x="gender", y="average_score", data=df)
plt.title("Average Score by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Score")
plt.show()
```

### Insight 3: Gender-wise Academic Performance

Female students show a slightly higher average academic score compared to male students.
This indicates potential differences in study patterns or learning approaches.


## Conclusion

This analysis explored how study hours, sleep duration, and gender relate to student academic performance.
The results show that higher study time and adequate sleep are associated with better scores.
Such insights can help students improve their academic outcomes through better time and health management.
